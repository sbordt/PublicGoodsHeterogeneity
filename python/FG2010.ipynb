# Fehr & GÃ¤chter 2010, Grouped-Fixed Effects

```python
import pandas
import cPickle as pickle
import numpy as np
import random

from gfe import gfe, gfe_bootstrap

path = "..\\data\\"

fg2010 = pandas.read_csv(path+"FG2010.csv", sep=',')
```

## Main regression

```python
_, betaHet, alpha, _, ass = gfe(fg2010[['cont']], None, fg2010[['belief', 'predictedcontribution']], 140, 10, 2, maxIter = 10000, hit = 15)
print betaHet

assignment = np.hstack((np.array(fg2010[['playerID']])[np.arange(1, 1400, 10), :], np.reshape(ass+1, (-1,1))))
np.savetxt(path+"FG2010_GFE.csv", assignment, fmt="%d, %d")

print np.mean(alpha[0,0:5])
print np.mean(alpha[0,5:10])

print np.mean(alpha[1,0:5])
print np.mean(alpha[1,5:10])

print np.sum(ass)
```

## Bootstrap replications (for the standard errors)

```python
random.seed(123)
np.random.seed(123)

_, betaHet_repl, ass_repl = gfe_bootstrap(fg2010[['cont']], None, fg2010[['belief', 'predictedcontribution']], 140, 10, 2, maxIter=2000, hit=15, nRepl=1000)
```

```python
pickle.dump(betaHet, open(path+"fg2010_betaHet.pkl", 'wb'), -1)
pickle.dump(alpha, open(path+"fg2010_alpha.pkl", 'wb'), -1)
pickle.dump(ass, open(path+"fg2010_ass.pkl", 'wb'), -1)
pickle.dump(betaHet_repl, open(path+"fg2010_betaHet_repl.pkl", 'wb'), -1)
pickle.dump(ass_repl, open(path+"fg2010_ass_repl.pkl", 'wb'), -1)
```
