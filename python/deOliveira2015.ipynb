# deOliveira et al. 2015, Grouped-Fixed Effects

```python
import pandas
import cPickle as pickle
import numpy as np
import random

from gfe import gfe, gfe_bootstrap

path = "..\\data\\"

data = pandas.read_csv(path+"deOliveira2015.csv", sep=',')
```

## Main regression

```python
_, betaHet, alpha, _, ass = gfe(data[['cont']], None, data[['belief']], data.shape[0]/15, 15, 2, maxIter = 10000, hit = 15)
```

```python
print betaHet
print alpha

print np.mean(alpha[0,0:5])
print np.mean(alpha[0,5:10])

print np.mean(alpha[1,0:5])
print np.mean(alpha[1,5:10])

print ass
print sum(ass)
print len(ass)
```

```python
assignment = np.hstack((np.array(data[['playerID']])[np.arange(1,data.shape[0],15),:],np.reshape(ass+1,(-1,1))))
np.savetxt(path+"deOliveira2015_GFE.csv", assignment, fmt="%d, %d")
```

## Bootstrap replications (for the standard errors)

```python
random.seed(123)
np.random.seed(123)

_, betaHet_repl, ass_repl = gfe_bootstrap(data[['cont']], None, data[['belief']], data.shape[0]/15, 15, 2, maxIter=10000, hit=15, nRepl=1000)
```

```python
pickle.dump(betaHet, open(path+"deOliveira_betaHet.pkl", 'wb'), -1)
pickle.dump(alpha, open(path+"deOliveira_alpha.pkl", 'wb'), -1)
pickle.dump(ass, open(path+"deOliveira_ass.pkl", 'wb'), -1)
pickle.dump(betaHet_repl, open(path+"deOliveira_betaHet_repl.pkl", 'wb'), -1)
pickle.dump(ass_repl, open(path+"deOliveira_ass_repl.pkl", 'wb'), -1)
```
